## Using matplotlib 


```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
G = 6.67430e-11  # Gravitational constant
c = 2.99792458e8  # Speed of light (m/s)
M_sun = 1.989e30  # Mass of the Sun (in kilograms)
R_sch = (2 * G * M_sun) / (c ** 2)  # Schwarzschild radius of the Sun

# Define the curvature function
def curvature(x, y):
    r = np.sqrt(x**2 + y**2)
    return 1 - R_sch / r

# Define the grid size and range
grid_size = 100
grid_range = np.linspace(-15, 15, grid_size)

# Create a meshgrid
X, Y = np.meshgrid(grid_range, grid_range)

# Compute the curvature at each point in the grid
curvature_values = curvature(X, Y)

# Create the figure
fig = plt.figure(figsize=(12, 6))

# Plot the 2D top view
ax1 = fig.add_subplot(121)
ax1.imshow(curvature_values, extent=[-25, 25, -25, 25], cmap='RdBu')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('Top View')
ax1.grid(True)

# Plot the 2D side view
ax2 = fig.add_subplot(122, projection='3d')
X_2d, Y_2d = np.meshgrid(grid_range, grid_range)
ax2.plot_surface(X_2d, Y_2d, curvature_values, cmap='RdBu')
ax2.set_xlabel('X')
ax2.set_ylabel('Z')
ax2.set_zlabel('Curvature')
ax2.set_title('Side View')

plt.suptitle('Space-Time Curvature around the Sun')

plt.tight_layout()
plt.show()
```

## Using Plotly

```python
import numpy as np
import plotly.graph_objects as go

# Constants
G = 6.67430e-11  # Gravitational constant
c = 2.99792458e8  # Speed of light (m/s)
M_sun = 1.989e30  # Mass of the Sun (in kilograms)
R_sch = (2 * G * M_sun) / (c ** 2)  # Schwarzschild radius of the Sun

# Define the curvature function
def curvature(x, y):
    r = np.sqrt(x**2 + y**2)
    return 1 - R_sch / r

# Define the grid size and range
grid_size = 100
grid_range = np.linspace(-20, 20, grid_size)

# Create a meshgrid
X, Y = np.meshgrid(grid_range, grid_range)

# Compute the curvature at each point in the grid
curvature_values = curvature(X, Y)

# Create a 2D surface plot using Plotly
fig = go.Figure(data=go.Surface(z=curvature_values, x=X, y=Y))
fig.update_layout(title='Space-Time Curvature around the Sun',
                  scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Curvature'))
fig.show()
```
